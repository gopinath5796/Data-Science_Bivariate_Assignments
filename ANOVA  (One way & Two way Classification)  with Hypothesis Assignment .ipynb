```python
import pandas as pd
```

```python
dataset=pd.read_csv("Pre_processed_Placement.csv")
dataset
```

# ANOVA [Analysis of Variance]:

# 1. One Way classification: One Independent Variable

```python
# 1.One Way classification:

import scipy.stats as stats

stats.f_oneway(dataset['ssc_p'],dataset['hsc_p'],dataset['degree_p'])
```

```python
# Result for One way classification

# H0= There is no significant between ssc_p, hsc_p & degree_p.
# H1= There is significant between ssc_p, hsc_p & degree_p.

# Here pValue=0.49 which is not < 0.05, So, Alternate Hypothesis is Rejected. 
# Finally, The Null Hypothesis is Accepted. There is no significant between ssc_p, hsc_p & degree_p.
```

# 2. Two Way Classification: Two Independent Variable

```python
# 2.Two Way Classification:

import statsmodels.api as sm
from statsmodels.formula.api import ols
anova_two=ols('sl_no ~ ssc_p + hsc_p + degree_p + ssc_p*hsc_p*degree_p',data=dataset).fit()
Two=sm.stats.anova_lm(anova_two,type=2)
Two
```

```python
# Result for Two Way classification

# Interaction = ssc_p * hsc_p * degree_P

# H0= There is no significant interaction effect between ssc_p * hsc_p * degree_P.
# H1= There is significant interaction effect between ssc_p * hsc_p * degree_P.

# Here pValue=0.714 which is not < 0.05, So, Alternate Hypothesis is Rejected. 
# Finally, The Null Hypothesis is Accepted. There is no significant interaction effect ssc_p * hsc_p * degree_P.
```
